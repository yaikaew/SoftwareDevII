```python
import sqlite3
from bs4 import BeautifulSoup

# Read the HTML file with the correct encoding
with open("dataQuerySelector.html", "r", encoding="cp874") as f:
    content = f.read()

# Use BeautifulSoup to parse the HTML content
soup = BeautifulSoup(content, "html.parser")

# Find the table in the HTML document
table = soup.find("table")

# Extract the rows and cells of the table
rows = table.find_all("tr")

# Connect to the database
conn = sqlite3.connect("db.sqlite3")
c = conn.cursor()

# Create the table if it doesn't exist
c.execute("CREATE TABLE IF NOT EXISTS courses (code TEXT, name TEXT, credit TEXT, date TEXT, time TEXT)")

# Loop through each row of the table and insert the data into the database
for row in rows:
    cells = row.find_all("td")
    if len(cells) == 4:
        print('1')
        code = cells[0].text.strip()
        name = cells[1].text.strip()
        credit = cells[2].text.strip()
        date_time = cells[3].text.strip().split()
        date = date_time[0]
        time = date_time[1]
        c.execute("INSERT INTO courses (code, name, credit, date, time) VALUES (?, ?, ?, ?, ?)", (code, name, credit, date, time))

# Commit the changes and close the connection
conn.commit()
conn.close()
```
